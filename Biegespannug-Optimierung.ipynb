```python
"""
PROJEKT: Automatisierter Biegespannungs-Rechner
AUTOR: Khaled Guedri
BESCHREIBUNG: Dieses Tool berechnet die Biegespannung eines Balkens und 
optimiert die Höhe automatisch, bis die Sicherheit gewährleistet ist.
"""

def main():
    print("--- Willkommen zum Maschinenbau-Dimensionierungstool ---")
    
    # 1. Eingaben
    try:
        breite = float(input("Breite des Balkens (mm): "))
        hoehe = float(input("Aktuelle Höhe des Balkens (mm): "))
        moment = float(input("Biegemoment (Nm): "))
    except ValueError:
        print("Fehler: Bitte nur Zahlen eingeben!")
        return

    # 2. Datenbank und Parameter
    materialien = {"S235": 235, "S355": 355, "Aluminium": 160}
    sicherheitsfaktor = 2.0
    
    wahl = input("Wähle Material (S235, S355, Aluminium): ")
    if wahl not in materialien:
        print("Material nicht gefunden.")
        return
    
    zulassige_spannung = materialien[wahl] / sicherheitsfaktor
    
    # 3. Optimierungsschleife (Iterative Berechnung)
    aktuelle_hoehe = hoehe
    spannung = (moment * 1000) / ((breite * aktuelle_hoehe**2) / 6)
    
    print(f"\nStarte Optimierung für {wahl}...")
    
    while spannung > zulassige_spannung:
        aktuelle_hoehe += 1  # Erhöhe Höhe um 1mm
        spannung = (moment * 1000) / ((breite * aktuelle_hoehe**2) / 6)
    
    # 4. Ergebnis-Ausgabe
    print("-" * 30)
    print(f"Ergebnis für Material {wahl}:")
    print(f"Optimierte Mindesthöhe: {aktuelle_hoehe} mm")
    print(f"Resultierende Spannung: {spannung:.2f} MPa")
    print(f"Zulässige Spannung: {zulassige_spannung:.2f} MPa")
    print("✅ Bauteil ist sicher dimensioniert.")

if __name__ == "__main__":
    main()
```
